```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:\MCA\DATA SCIENCE\DS LAB\Mall_Customers.csv")
df.head()
```

```python
plt.scatter(x=df['Annual Income (k$)'],y=df['Spending Score (1-100)'])
plt.show()
```

```python
model = KMeans(n_clusters=5)
df['Cluster'] = model.fit_predict(df[['Annual Income (k$)','Spending Score (1-100)']])
```

```python
plt.scatter(x=df['Annual Income (k$)'],y=df['Spending Score (1-100)'],c=df['Cluster'])
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],c='purple',marker="+",label="centroid")
```

```python
sse = []
for k in range(1,21):
    model = KMeans(n_clusters=k)
    model.fit_predict(df[['Annual Income (k$)','Spending Score (1-100)']])
    sse.append(model.inertia_)
sse
```

```python
plt.plot(range(1,21),sse)
```
